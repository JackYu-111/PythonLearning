```python
import pandas as pd
fpath='./Learning4_complex_students_2025.csv'
```

```python
df_students = pd.read_csv(fpath)
```

```python
df_students.head()
```

```python
df_students['Student_ID']
```

```python
#remove "SID-" of Student_ID for easier querying
df_students.loc[:,'Student_ID'] = df_students['Student_ID'].str.replace('SID-','').astype(object)
```

```python
df_students['Student_ID']
```

```python
from datetime import datetime
df_students['Birth_Date'] = pd.to_datetime(df_students['Birth_Date'])
```

```python
# calculate the age of students and return to new column 'Age'
current_date = datetime(2025,6,18)
df_students.loc[:,'Age'] = ((current_date - df_students['Birth_Date']).dt.days/365.25).astype(int)
```

```python
df_students[['First_Name','Last_Name','Age']]
```

```python
# df.apply for giving grade of English
def English_Grade(x):
    if x['English_Score'] > 90:
        return 'A'
    if x['English_Score'] > 70:
        return 'B'
    if x['English_Score'] > 50:
        return 'C'
    return 'F'
        
```

```python
df_students.loc[:,'English_Grade'] = df_students.apply(English_Grade,axis = 1)
```

```python
df_students[['Student_ID','English_Score','English_Grade']]
```

```python
def Math_Grade(a):
    if a > 90:
        return 'A'
    if a > 70:
        return 'B'
    if a > 50:
        return 'C'
    return 'F'
```

```python
df_students.loc[:,'Math_Grade'] = df_students['Math_Score'].apply(Math_Grade)
```

```python
df_students[['Student_ID','Math_Score','Math_Grade']]
```

```python
# re-arrange English_Grade,Math_Grade and Age columns
cols = df_students.columns.tolist()
print(cols)
```

```python
EG = cols.index('English_Grade')
cols.pop(EG)
cols.insert((cols.index('English_Score')+1),'English_Grade')
```

```python
MG = cols.index('Math_Grade')
cols.pop(MG)
cols.insert((cols.index('Math_Score')+1),'Math_Grade')
AG = cols.index('Age')
cols.pop(AG)
cols.insert((cols.index('Birth_Date')+1),'Age')
```

```python
df_students = df_students[cols]
```
