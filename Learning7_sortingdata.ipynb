```python
# Sorting data by ascending/descending
import pandas as pd
```

```python
fpath ='./complex_students_2025_2.csv'
```

```python
df =pd.read_csv(fpath)
```

```python
df
```

```python
#ascending of english score
df['English_Score'].sort_values()
```

```python
#descending of Math Score
df['Math_Score'].sort_values()
```

```python
#ascending of english score for whole data
df.sort_values(by ='English_Score')
```

```python
#ascending of english score, then Math score
df.sort_values(by =['English_Score','Math_Score'])
```

```python
#only Full Name, English Score and Math Score data with ascending of enlish score, then Math Score
df[['Full_Name','English_Score','Math_Score']].sort_values(by =['English_Score','Math_Score'])
```

```python
#only Full Name, English Score and Math Score data with ascending of enlish score, then descending of Math Score
df[['Full_Name','English_Score','Math_Score']].sort_values(by =['English_Score','Math_Score'], ascending =[True,False])
```

```python
df[['Last_Name','English_Score']].sort_values(by=['English_Score','Last_Name'],ascending =[True,False])
```

```python
df.sort_values(by=['English_Score','Last_Name'],ascending =[True,False])
```
